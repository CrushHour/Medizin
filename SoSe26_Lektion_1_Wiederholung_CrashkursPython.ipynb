## Aufgabe 4

def - Funktion definieren

![Quelle: Learn by example](https://www.learnbyexample.org/wp-content/uploads/python/Python-Function-Syntax.png)

Erstelle eine Funktion `berechne_preis`, die zwei Werte akzeptiert: den `original_preis` und den `rabatt_prozent`.

````python
# Funktion definieren
def berechne_preis(original_preis, rabatt_prozent):

    # Den Rabattbetrag ausrechnen
    rabattbetrag = original_preis * rabatt_prozent / 100
    neuer_preis_direkt = original_preis * (1-rabatt_prozent/100)

    # Den neuen Preis berechnen
    neuer_preis = original_preis - rabattbetrag
    if neuer_preis != neuer_preis_direkt:
        print("Achtung: Es gibt einen Fehler in der Berechnung!")
        return neuer_preis
    else:
        print("Die Berechnung ist korrekt.")
        return neuer_preis_direkt


    # Das Ergebnis "ausspucken"

# Die Funktion aufrufen
# Wir geben 100 (für original_preis) und 20 (für rabatt_prozent) hinein.
ergebnis = berechne_preis(100, 20)

# Das Ergebnis anzeigen
print("Der Preis nach Rabatt ist: {} Euro".format(ergebnis))
````

## Aufgabe 5

for loop

![Quelle: Learn by example](https://www.learnbyexample.org/wp-content/uploads/python/Python-for-Loop-Syntax.png)

Ein Patient im Krankenhaus ist an einen Monitor angeschlossen. Wir haben eine Liste mit den gemessenen Herzfrequenzen (Beats Per Minute - BPM) der letzten 5 Minuten.

Wir wollen jede Messung prüfen. Wenn ein Puls über 100 BPM liegt, soll das System einen Alarm ausgeben.

Liste mit Pulswerten:
```python
puls_werte = [72, 85, 110, 95, 125]
```
Aufgabe:
1. Schreibe eine `for`-Schleife, die jeden Wert in der Liste puls_werte nacheinander durchgeht:

2. Prüfe mit einem `if`, ob der aktuelle Puls größer als 100 ist.

    * Wenn ja, gib aus: "ALARM: Puls ist zu hoch! ([Wert] BPM)".

    * Wenn nein, gib aus: "Normaler Puls: ([Wert] BPM)".

````python
# Pulswerte definieren

puls_werte = [72, 85, 110, 95, 125]
# Die for-Schleife schreiben
for i in range(len(puls_werte)):

    # Die Bedingung prüfen mit if
    if puls_werte[i] > 100:
        print("ALARM: Pulswert ist zu hoch! ({} BPM)".format(puls_werte[i]))
    else:
        print("Pulswert ist normal: {} BPM".format(puls_werte[i]))
````

## Aufgabe 6

DataFrames

Du hast eine Liste von Patienten mit ihrem Namen, Alter und Blutdruck. Du sollst herausfinden, welche Patienten einen erhöhten Blutdruck haben.

1. Um mit Tabellen zu arbeiten, müssen wir zuerst die `pandas`-Bibliothek laden.
2. Erstelle einen DataFrame namens `df_patienten` mit folgenden Daten:

    * Name: "Müller", "Schmidt", "Fischer", "Weber"

    * Alter: 45, 62, 33, 71

    * Blutdruck_Syst: 120, 145, 118, 160

3. Filter den DataFrame so, dass nur Patienten angezeigt werden, deren Blutdruck_Syst über 140 liegt (Hypertonie Grad 1).

````python
# Richtige Bibliothek importieren
import pandas as pd

# Daten als Dictionary definieren
daten = {
    "Name": ["Müller", "Schmidt", "Fischer", "Weber"],
    "Alter": [45, 62, 33, 71],
    "Blutdruck_Syst": [120, 145, 118, 160]
}

# DataFrame erstellen
df_patienten = pd.DataFrame(daten)


# Daten filtern
# Wir sagen: Zeige mir alle Zeilen aus df_patienten, wo die Spalte Blutdruck_Syst > 140 ist
df_filtered = df_patienten[df_patienten["Blutdruck_Syst"] > 140]


# Ergebnis ausgeben
print("Patienten mit erhöhtem Blutdruck:")
print(df_filtered)
````
